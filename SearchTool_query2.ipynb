```python
store = pd.HDFStore('epri2.h5')
df = store['epri']
```

```python
def howmany(text,term,weight):
    return (text.count(term) * weight)

df["score"] = 0

def update_scores(row):
    scores = df["text"].apply(howmany, args=(row['term'],row['weight']))
    df["score"] = df["score"] + scores

grading_tool.apply(update_scores, axis=1)
```

## Result Set Using only query 2 all weighted '5'

```python
pd.set_option('display.max_colwidth', -1)
HTML(df.sort_values(by="score", ascending=False)[["Number","Program Name","Last Updated","url","score"]].to_html(escape=False, index=False))
```
